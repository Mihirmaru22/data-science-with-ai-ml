```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## basic ploting for seaborn

```python
tips = sns.load_dataset('tips')
tips
```

## scatter plot

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('scatterplot of total bill vs tips')
plt.show()
```

## line plot

```python
sns.lineplot(x='size',y='total_bill', data = tips)
plt.title('line plot based on size')
plt.show()
```

## catagorical plots
### bar plot

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('bar plot of total bill by day')
plt.show()
```

### blox plot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('boxplot')
plt.show() 
```

### violin plot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('violin plot')
plt.show() 
```

### histogram

```python
sns.histplot(data=tips,x='total_bill',bins=10, kde=True)
plt.title('histogram')
plt.show() 
```

### KDE plot

```python
sns.kdeplot(data=tips,x='total_bill', fill=True)
```

## pair plot

```python
sns.pairplot(tips)
```

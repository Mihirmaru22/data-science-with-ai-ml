maltiple plots


```python
x = [1, 2, 3, 4, 5, 6, 7]
sales = [120, 150, 170, 200, 180, 220, 210]
profit = [20, 25, 30, 35, 28, 40, 38]
plt.figure(figsize=(11,7))

plt.subplot(1,2,1)
plt.ylabel('sales')
plt.plot(x,sales, color='green')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.ylabel('profit')
plt.plot(x,profit, color='blue')
plt.title('plot 2')

plt.subplot(2,2,4)
plt.ylabel('sales')
plt.xlabel('profit')
plt.plot(profit,sales, color='red')
plt.title('plot 3')
```

### sales data visulization

```python
import pandas as pd
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

```python

df['Publisher'] = df['Publisher'].fillna('unknown')
df['Year'] = df['Year'].fillna(df['Year'].bfill().ffill())
df.info()
```

# total sales by publisher

```python
total_sale_by_product = df.groupby('Genre')['Global_Sales'].sum()
print(total_sale_by_product)
```

```python
total_sale_by_product.plot(kind='bar', color='lightskyblue')
```

### plot sales trend overtime


```python
sales_trend = df.groupby('Year')['Global_Sales'].sum().reset_index()

plt.plot(sales_trend['Year'],sales_trend['Global_Sales'])
```
